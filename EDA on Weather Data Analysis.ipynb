```python
plt.figure(figsize=(5, 8))
plt.hist(w_df['Rel Hum_%'], bins=20, color='red', alpha=0.7)
plt.xlabel('Humidity scale')
plt.ylabel('Frequency scale')
plt.title('Humidity scale Distribution')
plt.show()
```
